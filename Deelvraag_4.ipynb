```python
# import World Bank data
import wbdata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
from scipy import stats

```

```python
#SP.DYN.CONU.ZS      	Contraceptive prevalence, any methods (% of women ages 15-49)
#SP.DYN.CONM.ZS      	Contraceptive prevalence, modern methods (% of women ages 15-49)
#SP.MTR.1519.ZS      	Teenage mothers (% of women ages 15-19 who have had children or are currently pregnant)
#SP.UWT.TFRT         	Unmet need for contraception (% of married women ages 15-49)
#SP.DYN.SMAM.MA      	Mean age at first marriage, male
#SP.DYN.SMAM.FE      	Mean age at first marriage, female
wbdata.get_data("SP.DYN.CONU.ZS")
wbdata.get_data("SP.DYN.CONM.ZS")
wbdata.get_data("SP.MTR.1519.ZS")
wbdata.get_data("SP.UWT.TFRT")
```

```python
# Dataframe per category with all countries
CONU_all = pd.DataFrame(wbdata.get_data("SP.DYN.CONU.ZS"))
CONM_all = pd.DataFrame(wbdata.get_data("SP.DYN.CONM.ZS"))
MTR_all = pd.DataFrame(wbdata.get_data("SP.MTR.1519.ZS"))
CON_UNMET_all = pd.DataFrame(wbdata.get_data("SP.UWT.TFRT"))
```

```python
CONU_all['country_code'] = CONU_all.country.astype(str).str.slice(8,10)
CONU_all['country'] = CONU_all.country.astype(str).str.slice(23,-2)
CONU_all['indicator'] = CONU_all.indicator.astype(str).str.slice(35,-2)
CONU_all["date"] = CONU_all["date"].astype(int) 
CONU_all["value"] = CONU_all["value"].astype(float).round(2) 
CONU_all
```

```python
#import csv van 15 slechtse landen
CPIA_worst15 = pd.read_csv("CPIA_worst15.csv")
CPIA_worst15["date"] = CPIA_worst15["date"].astype(int) 
CPIA_worst15   
```

```python
CONU_2007_2017 = CONU_all.loc[(CONU_all['date'] >= 2007) & (CONU_all['date'] <= 2017)]
CONU_2007_2017
```

```python
CONU_worst15_alltime = CONU_all[CONU_all.country.isin(list(CPIA_worst15.country))]
CONU_worst15_2007_2017= CONU_2007_2017[CONU_2007_2017.country.isin(list(CPIA_worst15.country))]
```

```python
CONU_years = CONU_worst15_2007_2017.sort_values(by='date',ascending=True)
CONU_years
```

#### Look what datapoints are available for Somalia and take the most recent value

```python
CONU_all.loc[(CONU_all['country'] == 'Somalia') & (CONU_all['value'] >= 0)]
CONU_SOMALIA_2006 = CONU_all.value.loc[(CONU_all['country'] == 'Somalia') & (CONU_all['date'] == 2006)]
CONU_2006_SO = CONU_all.value.loc[(CONU_all['country'] == 'Somalia') & (CONU_all['date'] == 2006)].values

# Let the value from date 2006 be filled in for the years 2007-20017
CONU_years.loc[CONU_years['country'] == 'Somalia','value'] = CONU_2006_SO
CONU_years.loc[CONU_years['country']=='Somalia']
```

```python
# visualiseren van verandering per land

fig,ax = plt.subplots(figsize=(20, 10))

for name in CONU_years["country"].unique():
    ax.plot(CONU_years[CONU_years.country==name].date,CONU_years[CONU_years.country==name].value,label=name)

#change size and location of legend
ax.legend(loc=(1.1, 0.2), fontsize = 'xx-large')

#tekst op x en y-as
plt.xticks(rotation = 65, fontsize = 20)
plt.yticks(fontsize = 20)

#labels
plt.title("Change in Contraceptive prevalance, any methods (% of women ages 15-49)", fontsize = 30)
plt.xlabel("Year", fontsize = 25)
plt.ylabel("%", fontsize = 25)

# stapgrootte x-as 
```
