```python
#importeren van GDP data

GDP_all = pd.DataFrame(wbdata.get_data("NY.GDP.PCAP.CD"))
GDP_all['country'] = GDP_all.country.astype(str).str.slice(23,-2)
GDP_all['indicator'] = GDP_all.indicator.astype(str).str.slice(35,-2)
GDP_all['value'] = GDP_all.value.astype(float).round(1)
GDP_all["date"] = GDP_all.date.astype(str)
#GDP_all.loc[(GDP_all["value"] >= 0)]
GDP_all.head()

# Van geen enkel jaartal is de GDP van Somalia bekend. 
#GDP_all.loc[(GDP_all["country"] == "Yemen, Rep.")]
```

#### Missing values bij 2017, dus misschien dat laatst bekende data pakken. Desnoods is dat van 2016. Dat kan dan in de tekst verwerkt worden. Of nog ergens een dataset vandaan halen

```python
# Dataframe met de meest recente waarde van GDP van elk land.

GDP_now = GDP_all.sort_values("date", ascending = False).drop_duplicates(["country"])

# nieuw dataframe met de GDP van de 15 landen met de laagste equality rating 

GDP_worst15 = pd.merge( CPIA_worst15, GDP_now, how = "left", left_on = ["country", "date"], right_on = ["country", "date"])
GDP_worst15

# missing data!
```

```python
# plot van de GDP van de landen waar equality rating het laagste is

# array maken van de twee kolommen die nuttig zijn 
country_worst15 = GDP_worst15.country.values
value_worst15 = GDP_worst15.value_y.values

colors = ["Red", "Blue", "Yellow", "Green", "Pink", "Cyan"]

value_worst15[np.isnan(value_worst15)] = 0

plt.figure(figsize=(30, 10))
plt.bar(country_worst15, value_worst15, color = colors, width = 0.8, align='center')

plt.xticks(np.arange( len(country_worst15) ), country_worst15, rotation = 65, fontsize = 20)
plt.yticks(fontsize = 20)

plt.title("GDP per capita, PPP (US$) in 2017", fontsize = 30)
plt.xlabel("Country", fontsize = 25)
plt.ylabel("GDP", fontsize = 25)

# De waarde van het land bovenop bar
for (i, value) in enumerate (value_worst15):
    plt.text(i- 0.4 , value + 100 , "%.2f" % (value) , fontsize = 20)
```

```python
# Dataframe met data over de jaren heen van de 15 landen

GDP_years = GDP_all.loc[((GDP_all["country"].isin(CPIA_worst15["country"]))),:].sort_values(by = "date", ascending = True)
GDP_years["date"] = GDP_years["date"].astype(int)

#drop rows with date lower dan 2000
GDP_lowerYears = GDP_years[GDP_years["date"] < 2000]
GDP_years = GDP_years.drop(GDP_lowerYears.index, axis = 0)

#show partial dataframe
GDP_years.head()
```

```python
# visualiseren van verandering per land

fig,ax = plt.subplots(figsize=(20, 10))

for name in GDP_years["country"].unique():
    ax.plot(GDP_years[GDP_years.country==name].date,GDP_years[GDP_years.country==name].value,label=name)

#change size and location of legend
ax.legend(loc=(1.1, 0.2), fontsize = 'xx-large')

#tekst op x en y-as
plt.xticks(rotation = 65, fontsize = 20)
plt.yticks(fontsize = 20)

#labels
plt.title("Change in GDP per capita (US$)", fontsize = 30)
plt.xlabel("Year", fontsize = 25)
plt.ylabel("GDP", fontsize = 25)

# stapgrootte x-as 

plt.xticks(range(2000, 2018))
```

### Extrapoleren

```python
import pandas as pd
from io import StringIO
from scipy.optimize import curve_fit

# Do the original interpolation
GDP_years.interpolate(method='nearest', xis=0, inplace=True)

# Display result
# print ('Interpolated data:')
# print (GDP_years)

# Function to curve fit to the data
def func(x, a, b):
    return a * (x) + b 

# Initial parameter guess, just to kick off the optimization
guess = (1, 1)

# Create copy of data to remove NaNs for curve fitting
fit_GDP_years = GDP_years.dropna()

# Place to store function parameters for each column
col_params = {}

# Curve fit each column
for col in fit_GDP_years.columns:  
    # Get x & y
    x = fit_GDP_years.index.astype(float).values
    y = fit_GDP_years[col].values
    # Curve fit column and get curve parameters
    params = curve_fit(func, x, y, guess)
    # Store optimized parameters
    col_params[col] = params[0]

# Extrapolate each column
for col in GDP_years.columns:
    # Get the index values for NaNs in the column
    x = GDP_years[pd.isnull(GDP_years[col])].index.astype(float).values
    # Extrapolate those points with the fitted function
    GDP_years[col][x] = func(x, *col_params[col])

# Display result
print ('Extrapolated data:')
print (GDP_years)

print ('Data was extrapolated with these column functions:')
for col in col_params:
    print ('f_{}(x) = {:0.3e} x + {:0.4f}'.format(col, *col_params[col]))
```
