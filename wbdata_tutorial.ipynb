```python
#Get data on CPIA gender equility rating from country Nepal(NPL)
wbdata.get_data("IQ.CPA.GNDR.XQ", country="NPL")
```

```python
# Steps to create DataFrame
CPIA_NPL = pd.DataFrame(wbdata.get_data("IQ.CPA.GNDR.XQ", country="NPL"))
CPIA_NPL
#
```

```python
#Clean the column value of 'country' by using str.slice
CPIA_NPL['country'] = CPIA_NPL.country.astype(str).str.slice(23,-2)
CPIA_NPL
```

```python
#Clean the column value of 'indicator' by using str.slice
CPIA_NPL['indicator'] = CPIA_NPL.indicator.astype(str).str.slice(35,-2)
CPIA_NPL
```

```python
# Making the column values of type 'float' and the values 'None' into NaN
CPIA_NPL['value'] = CPIA_NPL.value.astype(float)
CPIA_NPL
```

```python
# Creating new name 'CPIA_NPL_value' that:
# Selects row 'value' based on bigger or equal to zero (thus NaN values are excluded)
# Is sorted in ascended order on column 'date'
CPIA_NPL_value = CPIA_NPL[ CPIA_NPL.value >= 0].sort_values(by='date', ascending=True )
CPIA_NPL_value
```

```python
# Plotting simple bar plot
CPIA_NPL_value.plot.bar(x='date',y='value')
plt.xlabel("CPIA Gender Equality rating in Nepal (1=low to 6=high))")
plt.ylabel("Rating")
```
